```python
#@title Working with Python Dictionaires
"""
Lab: Working with Python Dictionaries

Objectives:
- Create and manipulate dictionaries in Python.
- Perform common dictionary operations like adding, updating, and removing key-value pairs.
- Understand dictionary methods and their applications.

Tasks:
Each task's result is stored in a variable, which is tested for correctness.
"""

# Task 1: Creating and Accessing Dictionary Elements
# Declare a dictionary representing a student with some basic information
student_info = {
    "name": "Alex Johnson",
    "age": 21,
    "major": "Computer Science",
    "graduated": False
}

# Task 1.1: Access the student's name and store it in a variable
student_name = student_info["name"]
print(f"The student's name is: {student_name}")

# Task 1.2: Access the student's age and store it in a variable
student_age = student_info["age"]
print(f"The student's age is: {student_age}")

# Task 2: Adding and Updating Dictionary Elements
# Task 2.1: Add a new key-value pair for the student's GPA. Key is GPA. And value is 3.8
student_info['GPA'] = 3.8
print(f"Updated student information with GPA: {student_info}")

# Task 2.2: Update the student's age to 22
student_info['age'] = 22
print(f"Updated student's age to: {student_info['age']}")

# Task 2.3: Update the 'graduated' status to True
student_info['graduated'] = True
print(f"Graduated status updated to: {student_info['graduated']}")

# Task 3: Removing Elements from a Dictionary
# Task 3.1: Remove the 'major' key from the dictionary
student_major =  student_info.pop("major")
print(f"Removed major: {student_major}")

# Task 4: Dictionary Methods
# Task 4.1: Get all keys of the dictionary
keys = list(student_info.keys())
print(f"Keys in the dictionary: {keys}")

# Task 4.2: Get all values of the dictionary
values = list(student_info.values())
print(f"Values in the dictionary: {values}")

# Task 4.3: Get all items (key-value pairs) of the dictionary
items = list(student_info.items())
print(f"Items in the dictionary: {items}")

# Task 4.4: Check if 'GPA' is in the dictionary keys
has_gpa = "GPA" in student_info
print(f"Is 'GPA' a key in the dictionary? {has_gpa}")

print("\nAll done! Testing now...\n\n")

# ======== Stop =================
# Do not change any code below ============
# The following code is used to test
# =========================================

# Keep track of marks
marks = 0

# Custom Tests and Marks Calculation
def check_variable(variable, expected_value, var_name):
    global marks
    if variable == expected_value:
        marks += 1
    else:
        print(f"{var_name} did not pass.")

# Run custom tests and add marks
check_variable(student_name, "Alex Johnson", "student_name")
check_variable(student_age, 21, "student_age")
check_variable(student_info.get("GPA"), 3.8, "student_info['GPA']")
check_variable(student_info.get("age"), 22, "student_info['age']")
check_variable(student_info.get("graduated"), True, "student_info['graduated']")
check_variable(student_major, "Computer Science", "student_major")
check_variable(keys, ["name", "age", "graduated", "GPA"], "keys")
check_variable(values, ["Alex Johnson", 22, True, 3.8], "values")
check_variable(items, [("name", "Alex Johnson"), ("age", 22), ("graduated", True), ("GPA", 3.8)], "items")
check_variable(has_gpa, True, "has_gpa")


# Final marks output
print(f"\nFinal Marks: {marks}")
```
