```python
# Task 1: Creating and Manipulating Strings
# Declare a string
text = "welcome to the world of Python!"

# Task 1.1: Print the length of the string
length_of_text = len(text)
print(f"The length of the string is: {length_of_text}")

# Task 1.2: Accessing characters
first_character =  text[0]
print(f"The first character is: {first_character}")
last_character =  text[-1]
print(f"The last character is: {last_character}")

# Task 2: String Slicing
# Task 2.1: Extract a substring "welcome"
welcome_word =  text[0:7]
print(f"The welcome word is: {welcome_word}")

# Task 2.2: Reverse the string
reversed_text =  text[::-1]
print(f"The reversed string is: {reversed_text}")

# Task 2.3: Skipping characters
every_second_char =  text[::2]
print(f"The string with every second character is: {every_second_char}")

# Task 3: Common String Methods
# Task 3.1: Converting to uppercase
uppercase_text =  text.upper()
print(f"The uppercase string is: {uppercase_text}")

# Task 3.2: Converting to lowercase
lowercase_text = text.lower()
print(f"The lowercase string is: {lowercase_text}")

# Task 3.3: Replace substring "Python" with "programming"
replaced_text = text.replace("Python","programming")
print(f"The replaced string is: {replaced_text}")

# Task 3.4: Splitting text spring by space
words = text.split()
print(f"The list of words is: {words}")

# Task 3.5: Joining strings by hyphen
hyphen_joined_text = "-".join(words)
print(f"The hyphen joined string is: {hyphen_joined_text}")

# Task 4: Advanced String Operations
# Task 4.1: Check for substring
contains_python = text.find("Python") != -1
print(f"Does the string contain 'Python'? {contains_python}")

# Task 4.2: Count occurrences
count_o = text.count('o')
print(f"The letter 'o' occurs {count_o} times in the string.")

# Task 4.3: Find substring index
index_of_world = text.find('world')
print(f"The index of the word 'world' is: {index_of_world}")

# Task 4.4: Find the charater at the index 4
char_at_index_4 = text[4]
print(f"The character at the index 4 is: {char_at_index_4}")


print("\n All done! Over to testing now.....!\n\n\n")

#======== Stop  =================
# Do not change any code below
# The following code is used to test
#==========================================

#keep track of marks
marks = 0

# Custom Tests and Marks Calculation
def check_variable(variable, expected_value, var_name):
    global marks
    if variable == expected_value:
        marks += 1
    else:
        print(f"{var_name} did not pass.")


# Run custom tests and add marks
check_variable(length_of_text, 31, "length_of_text")
check_variable(first_character, 'w', "first_character")
check_variable(last_character, '!', "last_character")
check_variable(welcome_word, "welcome", "welcome_word")
check_variable(reversed_text, "!nohtyP fo dlrow eht ot emoclew", "reversed_text")
check_variable(every_second_char, "wloet h ol fPto!", "every_second_char")
check_variable(uppercase_text, "WELCOME TO THE WORLD OF PYTHON!", "uppercase_text")
check_variable(lowercase_text, "welcome to the world of python!", "lowercase_text")
check_variable(replaced_text, "welcome to the world of programming!", "replaced_text")
check_variable(words, ['welcome', 'to', 'the', 'world', 'of', 'Python!'], "words")
check_variable(hyphen_joined_text, "welcome-to-the-world-of-Python!", "hyphen_joined_text")
check_variable(contains_python, True, "contains_python")
check_variable(count_o, 5, "count_o")
check_variable(index_of_world, 15, "index_of_world")
check_variable(char_at_index_4, 'o', "char_at_index_4")


# Final marks output
print(f"\nFinal Marks: {marks}")
```
